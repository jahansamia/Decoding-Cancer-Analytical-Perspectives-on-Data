```python
cancer_data = pd.read_csv('/content/cancer-issue/cancer issue.csv')

print(cancer_data.head())
```

```python
print(cancer_data.info())
```

```python
print(cancer_data.describe())
```

```python
# 1. Cancer Type by Gender:
fig2 = px.histogram(cancer_data, x="CancerType", color="Gender", barmode="group",
                   title="Cancer Type by Gender")
fig2.update_xaxes(tickangle=45, tickmode='linear')
fig2.show()
```

#### Cancer Type by Gender:
 The histogram likely reveals variations in cancer type prevalence between genders (e.g., Breast cancer may be more common in females)

```python
# 2. Age Distribution by Cancer Type:
plt.figure(figsize=(10, 6))
sns.boxplot(x='CancerType', y='Age', data=cancer_data)
plt.title('Age Distribution by Cancer Type')
plt.xlabel('Cancer Type')
plt.ylabel('Age')
plt.xticks(rotation=45, ha='right')
plt.show()
```

####Age Distribution:
 The box plots likely show age distributions for different cancer types and stages, potentially highlighting age-related cancer trends.

```python
# 3. Age Distribution by Cancer Type and Stage:
fig1 = px.box(cancer_data, x="CancerType", y="Age", color="Stage",
              title="Age Distribution by Cancer Type and Stage")
fig1.update_xaxes(tickangle=45, tickmode='linear')  # Rotate x-axis labels
fig1.show()
```

#### Age distribution by Cancer type and stage:
This visualizes the relationship between age, cancer type, and cancer stage. Each box in the plot represents the distribution of age for a specific cancer type and stage. This can help identify trends and patterns, such as whether certain cancer types are more prevalent in specific age groups or if age at diagnosis varies by cancer stage.

```python
# 4. Survival Month vs Treatment Response:
fig3 = px.violin(cancer_data, x="TreatmentResponse", y="SurvivalMonths", color="CancerType",
                 box=True, points="all", title="Survival Months by Treatment Response and Cancer Type")
fig3.show()
```

#### Survival by Treatment Response:
 The violin plot likely illustrates the relationship between treatment response, survival months, and cancer type, providing insights into treatment effectiveness.

```python
# 5. Distribution of Survival Months by Cancer Type:
fig4 = px.histogram(cancer_data, x="SurvivalMonths", color="CancerType",
                   marginal="rug",  # Add a marginal rug plot
                   title="Distribution of Survival Months by Cancer Type")
fig4.show()
```

#### Survival by Treatment Response:
 The violin plot likely illustrates the relationship between treatment response, survival months, and cancer type, providing insights into treatment effectiveness.

```python
# 6. Tumor size by Cancer Type
plt.figure(figsize=(10, 6))
sns.histplot(data=cancer_data, x='TumorSize', hue='CancerType', element='step',
             kde=True, common_norm=False)
plt.title('Distribution of Tumor Size by Cancer Type')
plt.xlabel('Tumor Size')
plt.ylabel('Frequency')
plt.show()
```

#### Tumor Size Distribution:
The histplot (with KDE) likely displays the distribution of tumor sizes for different cancer types, aiding in understanding tumor size variability.

```python
# 7. BMI Distribution by Smoking Status and Family History:
fig5 = px.violin(cancer_data, x="SmokingStatus", y="BMI", color="FamilyHistory",
                 box=True, points="all",  # Show box plot and individual points
                 title="BMI Distribution by Smoking Status and Family History")
fig5.show()
```

#### BMI, Smoking, and Family History:
 The violin plot likely explores the relationship between BMI, smoking status, and family history, potentially suggesting risk factor associations.

```python
# 8. Correlation Matrix:
plt.figure(figsize=(12, 8))
numerical_features = cancer_data.select_dtypes(include=['number'])
sns.heatmap(numerical_features.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

####Correlation Matrix:
 The heatmap likely reveals correlations between numerical features, providing insights into relationships between factors like age, BMI, tumor size, and survival..
